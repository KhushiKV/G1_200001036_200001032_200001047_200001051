```python
#importing dataset
train_dir = "./train_test_mails/train-mails" # Training dataset folder
test_dir = "./train_test_mails/test-mails" # Testing dataset folder
```

```python
# Creating a list containing all the words present in the training set emails
emails = [os.path.join(train_dir,f) for f in os.listdir(train_dir)] 
emails
```

```python
#creating an array of all the words present in the messages
all_words = []
for mail in emails:
    with open(mail) as m:
        for i, line in enumerate(m):
            if i==2:  #since message body is present in the second line in our dataset mails.
                words = line.split()
                all_words += words
```

```python
# Create a dictionary consisting of the freqeuncy of the words occurring in the trainset using Counter class
from collections import Counter
dictionary = Counter(all_words)

dictionary
```

```python
#filtering the dictionary to keep only those words that are not stopwords and are proper english strings with length greater than 1 atleast.
from nltk.corpus import stopwords

to_remove = list(dictionary.keys())
for item in to_remove:
    if item.isalpha() == False:
        del dictionary[item]
    elif len(item) == 1:
        del dictionary[item]
    elif item in stopwords.words('english'):
        del dictionary[item]
```

```python
dictionary
```

```python
# Generating a list of most common 3000 words in the training emails
dictionary = dictionary.most_common(3000)
```

```python
# Assigning an index to each word in the vocabulary
vocab = {}
for i, (word, frequency) in enumerate(dictionary):
    vocab[word] = i

vocab
```

```python
#Feature Extraction:
#We will create a matrix, that will contain eac row for one text message,each column for the words in the dictionary, 
#and data cels entry will tell the frequency of that word in that particular text message.

import numpy as np
def extract_features(mail_dir, vocab):
    files = [os.path.join(mail_dir, fi) for fi in os.listdir(mail_dir)]
    # print(files)
    docId = 0
    matrix = np.zeros((len(files),3000))
    for file in files:
        with open(file) as m:
            for i, line in enumerate(m):
                # print(line)
                if i==2:
                    words = line.split() # Converting to tokens
                    wordId = 0
                    for word in words:
                        # print(word)
                        if word not in vocab.keys():
                            continue
                        wordId = vocab[word]
                        # print(docId, wordId)
                        matrix[docId, wordId] = words.count(word)  #filling the matrix with values
            docId += 1
    
    return matrix
```

```python
word_matrix = extract_features(train_dir, vocab)
```

```python
#creating train labels for training dataset
train_labels = np.zeros(702) # Training dataset
train_labels[351:701] = 1    # Training Labels
```

```python
#Creating and Training the Model:

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, accuracy_score

model = MultinomialNB()
model.fit(word_matrix,train_labels) #Fit the Naive Bayes' Model
```

```python
#Preparing the test dataset for testing on model

test_labels = np.zeros(260)
test_labels[130:260] = 1
test_word_matrix = extract_features(test_dir, vocab)
```

```python
#Testing and making preditions on the Test dataset:

predictions = model.predict(test_word_matrix)
```

```python
#printing confusion matrix
print(confusion_matrix(test_labels, predictions))
```

```python
#printing accuracy_score
print(accuracy_score(test_labels,predictions))
```
